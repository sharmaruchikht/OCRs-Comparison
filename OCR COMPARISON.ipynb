### EASYOCR

```python
import easyocr
import re
import urllib.request
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Load the EasyOCR reader
reader = easyocr.Reader(['en'])

# Load the image from URL
image_url = 'https://img.freepik.com/premium-vector/food-flyer-template_747363-13.jpg'
urllib.request.urlretrieve(image_url, "a.jpg")

# Perform OCR on the image
result = reader.readtext("a.jpg")

# Display the image
img = mpimg.imread("a.jpg")
plt.imshow(img)
plt.axis('off')
plt.show()

# Print the extracted text
for item in result:
    print(item[1])
```

### PADDLEOCR

```python
from paddleocr import PaddleOCR, draw_ocr
from matplotlib import pyplot as plt
import cv2
import os
import numpy as np
import urllib.request

# Initialize the PaddleOCR model
ocr_model = PaddleOCR(lang='en')

# Load the image from the URL into memory
url = 'https://img.freepik.com/premium-vector/food-flyer-template_747363-13.jpg'
with urllib.request.urlopen(url) as url_response:
    img_array = np.asarray(bytearray(url_response.read()), dtype=np.uint8)

# Read the image using OpenCV
img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)

# Run the OCR method on the PaddleOCR model
result = ocr_model.ocr(img)

# Display the image
img = mpimg.imread("a.jpg")
plt.imshow(img)
plt.axis('off')
plt.show()

# Print the extracted text
for line in result:
    for res in line:
        print(res[1][0])
```

### GOOGLE OCR API

```python
# pip install google-cloud-vision
```

```python
import os, io
from google.cloud import vision
```

```python
def extract_text_from_image_url(credentials_path, image_url):
    os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = credentials_path
    client = vision.ImageAnnotatorClient()

    image = vision.Image()
    image.source.image_uri = image_url

    response = client.text_detection(image=image)
    texts = response.text_annotations

    if texts:
        extracted_text = texts[0].description
    else:
        extracted_text = ''

    return extracted_text

# Replace 'YOUR_CREDENTIALS_PATH' with the path to your Google Cloud credentials JSON file
credentials_path = 'OCR_APIs/keys/...'

# Replace 'IMAGE_URL' with the URL of the image you want to extract text from
image_url = 'https://img.freepik.com/premium-vector/food-flyer-template_747363-13.jpg'

# Display the image
img = mpimg.imread("a.jpg")
plt.imshow(img)
plt.axis('off')
plt.show()

# Extract text from the image
extracted_text = extract_text_from_image_url(credentials_path, image_url)
print(extracted_text)
```
